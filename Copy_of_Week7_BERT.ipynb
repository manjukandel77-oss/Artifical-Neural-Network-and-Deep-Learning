```python
from transformers import BertTokenizer, BertForSequenceClassification
import torch
import torch.nn.functional as F
```

```python
model_name = "nlptown/bert-base-multilingual-uncased-sentiment"

tokenizer = BertTokenizer.from_pretrained(model_name)
model = BertForSequenceClassification.from_pretrained(model_name)
```

```python
def predict_sentiment(text):
    # Tokenize input text
    inputs = tokenizer(text, return_tensors="pt", truncation=True, padding=True)

    # Get output from model
    outputs = model(**inputs)

    # Convert logits to probabilities
    probs = F.softmax(outputs.logits, dim=1)

    # Get predicted class
    predicted_class = torch.argmax(probs).item()

    return predicted_class + 1, probs
```

```python
text = "Ncell internet is very slow and customer service is disappointing."

rating, probability = predict_sentiment(text)

print("Input Comment:", text)
print("Predicted Rating (1 = Negative, 5 = Positive):", rating)
print("Probabilities:", probability)
```

```python
comments = [
    "Ncell network is down again, very bad service!",
    "The internet speed is okay, but could be better.",
    "Great service! I am happy with Ncell.",
    "Worst customer support experience ever.",
    "Thank you Ncell for fast internet today!"
]

for c in comments:
    rating, _ = predict_sentiment(c)
    print(f"Comment: {c}")
    print(f"Sentiment Rating: {rating} ⭐\n")
```

```python
comment = input("Enter your comment: ")

# Unpack the tuple to get the rating score. The second element (probabilities) is not used in the subsequent code.
rating, _ = predict_sentiment(comment)

print("\n-----------------------------")
print("Your Comment:", comment)
print("Predicted Sentiment Rating:", rating, "⭐")
print("-----------------------------")

if rating <= 2:
    print("Sentiment Result: NEGATIVE ❌")
elif rating == 3:
    print("Sentiment Result: NEUTRAL ⚠️")
else:
    print("Sentiment Result: POSITIVE ✅")
```
