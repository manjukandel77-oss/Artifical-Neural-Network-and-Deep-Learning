```python
# ==========================================
# MATRIX BACKPROPAGATION (3 -> 3 -> 1)
# ==========================================

import numpy as np

# Input (1x3) and Target (1x1)
X = np.array([[2.0, 1.0, 3.0]])
y = np.array([[6.0]])

# Learning Rate
lr = 0.05

# New Initial Weights (changed again)
W1 = np.array([[0.9, -0.4, 0.1],
               [-0.2, 0.6, 0.3],
               [0.5, -0.7, 0.8]])

W2 = np.array([[0.3],
               [1.2],
               [-0.6]])

print("\n=== INITIAL WEIGHTS ===")
print("W1:\n", W1)
print("W2:\n", W2)

# ReLU functions
def relu(z):
    return np.maximum(0, z)
def relu_derivative(z):
    return (z > 0).astype(float)

# ==========================================
# FORWARD PASS
# ==========================================

Z1 = np.dot(X, W1)
H1 = relu(Z1)
y_pred = np.dot(H1, W2)

print("\n=== FORWARD PASS ===")
print("Prediction:", round(y_pred[0][0], 4))
print("Target:", y[0][0])

# ==========================================
# LOSS CALCULATION
# ==========================================

dy_pred = y_pred - y
loss = 0.5 * np.sum(dy_pred ** 2)

print("\n=== LOSS CALCULATION ===")
print("Error:", round(dy_pred[0][0], 4))
print("MSE Loss:", round(loss, 4))

# ==========================================
# BACKPROPAGATION
# ==========================================

dW2 = np.dot(H1.T, dy_pred)
dH1 = np.dot(dy_pred, W2.T)
dZ1 = dH1 * relu_derivative(Z1)
dW1 = np.dot(X.T, dZ1)

print("\n=== GRADIENTS ===")
print("dW1:\n", dW1)
print("dW2:\n", dW2)

# ==========================================
# WEIGHT UPDATE
# ==========================================

W1_new = W1 - lr * dW1
W2_new = W2 - lr * dW2

print("\n=== UPDATED WEIGHTS ===")
print("New W1:\n", np.round(W1_new, 4))
print("New W2:\n", np.round(W2_new, 4))

# ==========================================
# FINAL CHECK
# ==========================================

new_Z1 = np.dot(X, W1_new)
new_H1 = relu(new_Z1)
new_prediction = np.dot(new_H1, W2_new)

print("\n=== FINAL RESULT ===")
print("Old Prediction:", round(y_pred[0][0], 4))
print("New Prediction:", round(new_prediction[0][0], 4))
print("If successful, the new prediction should be closer to", y[0][0])
```
